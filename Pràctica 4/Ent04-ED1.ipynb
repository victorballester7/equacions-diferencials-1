#### Exercici 1

```python
var('c k y1')
```

```python
y=function('y')(x)
```

```python
# feix de corbes
f1= x^4+y^2==c
```

```python
diff(f1,x) # diferenciem implícitament
```

```python
# calculem la derivada dy/dx en cada punt del feix de corbes.
g=solve(diff(f1,x),diff(y(x), x))
g
```

```python
g1=g[0].rhs()
g1 # g1 = dy/dx
```

```python
# plantejem l'equació diferencial: el pendent (la derivada) de la solució ha de ser -1/g1, perquè volem que les
# corbes solucions de l'equació diferencial siguin ortogonals al feix de corbes anterior.
edo1= diff(y(x), x)== - 1/g1
```

```python
# solucionem l'equació diferencial i canviem la constant _C per una constant k (per tal d'evitar errors més
# endavant).
sol1=desolve(edo1,dvar=y,ivar=x).subs(_C=k)
sol1
```

```python
# Per tant, teòricament la família de corbes ortogonals al feix donat són k*e^(-1/4/x^2) amb k un número real.
# Comprovem-ho ara gràficament
```

```python
# Canviem la funció y(x) per una variable auxiliar y1 per tal de no tenir problemes a l'hora de graficar la funció.
ff1=f1.subs(y(x)==y1)
ff1
```

```python
# feix de corbes per a c=1,2,3,4,5,6.
plot1=[implicit_plot(ff1.subs(c=i),(x,-2,2),(y1,-3,3),color='red') for i in range(1,7)]
sum(plot1)
```

```python
# feix de solucions de l'equació diferencial per a k=-5,-4.25,-3.5,...4.75
L1=[plot(sol1.subs(k=i),-3,3,ymin=-5,ymax=5) for i in srange(-5,5,0.75)]
sum(L1)
```

```python
L11=[plot(sol1.subs(k=i),-3,3,ymin=-3,ymax=3) for i in srange(-5,5,0.75)]
sum(plot1)+sum(L11)
```

```python
# Gràficament podem observar que els talls són, en efecte, ortogonals.
L111=[plot(sol1.subs(k=i),0,1,ymin=0.5,ymax=2.6) for i in srange(-5,5,0.75)]
plot111=[implicit_plot(ff1.subs(c=i),(x,0,1),(y1,0.5,2.6),color='red') for i in range(1,7)]
sum(plot111)+sum(L111)
```

#### Exercici 3

```python
# Considerem la reacció A + B + C --> D
# Suposem que per formar 10 grams de D es necessiten 6 de A, 3 de B i 1 de C.
```

```python
var('t k b A0 B0 C0')
```

```python
A=function('A')(t)
B=function('B')(t)
C=function('C')(t)
D=function('D')(t)
```

```python
# Com a condicions inicials prenem els següents valors:
A0=100
B0=40
C0=20
```

```python
# Equacions diferencials que modelitzen el nostre problema:
edo31= diff(A,t)==-6*k*A*B*C
edo32= diff(B,t)==-3*k*A*B*C
edo33= diff(C,t)==-1*k*A*B*C
edo34= diff(D,t)==10*k*A*B*C

edos=[edo31,edo32,edo33]
```

```python
# De la manera com hem definit les equacions diferencials, k ha de ser positiu.
assume(k>0)
```

```python
# Calculem simultàniament A(t), B(t) i C(t) en funció de D(t).
S=[] 
M=[A,B,C]
for i in range(len(edos)):
    s=desolve(edo34/edos[i],D,ivar=t,ics=[0,0],contrib_ode=True) # Resolem l'equació diferencial (per exemple 
    # pel cas de A) diff(D,t)/diff(A,t)=-10/6.
    sol=solve(D==s,M[i])[0].rhs()
    S.append(sol)
S
```

```python
# substituim les condicions inicials.
S[0]=S[0].subs(A(0)==A0)
S[1]=S[1].subs(B(0)==B0)
S[2]=S[2].subs(C(0)==C0)
S
```

```python
# substituim aquestes equacions a l'equació diferencial de D(t).
edo344=edo34.subs(A(t)==S[0],B(t)==S[1],C(t)==S[2])
edo344
```

```python
# resolem la equació diferencial de D(t).
sol=desolve(edo344,D,ivar=t,ics=[0,0])
show(sol)
```

```python
# simplifiquem l'expressió.
sol2=sol.simplify_log()*400*k
show(sol2)
```

```python
# aïllem D(t).
solD=solve(sol2,D)
show(solD)
```

```python
# Vegem quina solució és l'adecuada. Sabem que D(0)=0. Per tant, comprovem quant valen aquestes dues funcions al 0.
print(solD[0](t=0))
print()
print(solD[1](t=0))
```

```python
# Per tant, la solució adecuada és la primera.
D1=solD[0].rhs()
```

```python
# Com que D(t) és creixent (per la definció de D(t)), el seu valor màxim l'aconseguirà en t=infinit. Per tant, el
# valor màxim de D(t) és:
limit(D1,t=oo)
```

```python
# Calculem ara les quantitats de A, B i C en un temps molt llarg. Per això, recuperem les relacions entre 
# A, B, C i D.
A1=S[0].subs(D(t)==D1)
B1=S[1].subs(D(t)==D1)
C1=S[2].subs(D(t)==D1)
```

```python
limit(A1,t=oo)
```

```python
limit(B1,t=oo)
```

```python
limit(C1,t=oo)
```

```python
# Per tant, B és el reactiu limitant.
```

```python
# Per tal de representar la solució gràficament, ens cal calcular un valor de k. Aquest l'obtindrem imposant la
# següent condició:
# "En 1 segon s'han format 50 grams de D"
```

```python
# Trobem doncs el valor de k que satisfà aquesta propietat numèricament.
K=find_root(D1.subs(t=1)==50,0,1) # Busquem un valor entre 0 i 1.
K
```

```python
pa=plot(A1.subs(k=K),0,20,color="green",legend_label='A')
pb=plot(B1.subs(k=K),0,20,color="purple",legend_label='B')
pc=plot(C1.subs(k=K),0,20,color="red",legend_label='C')
pd=plot(D1.subs(k=K),0,20,color="blue",legend_label='D')
pa+pb+pc+pd
```

```python
# Ara augmenten la quantitat de l'element més restrictiu. Aquest element és el B, ja que quan s'agota (en un temps 
# infinit) la reacció s'acaba. A més observem que el segon element més restrictiu és A, ja que 
# A0/6 = 100/6 < 20/1 = C0/1.

# Suposem que B0=40+r, on r és un paràmetre positiu. A mesura que anem augmentant r, limit(A,t=oo) anirà disminuint
# fins arribar a 0 i a partir de llavors el nou reactiu limitant passarà a ser A. Per tant, hi haurà un valor de r
# tal que limit(A,t=oo)=limit(B,t=oo)=0. Calculem-lo!
var('r')
B0=40+r
B0
```

```python
# Calculem simultàniament A(t), B(t) i C(t) en funció de D(t) i substituïm les noves condicions inicials (com he 
# fet anteriorment).
S=[]
M=[A,B,C]
for i in range(len(edos)):
    s=desolve(edo34/edos[i],D,ivar=t,ics=[0,0],contrib_ode=True)
    sol=solve(D==s,M[i])[0].rhs()
    S.append(sol)
S[0]=S[0].subs(A(0)==A0)
S[1]=S[1].subs(B(0)==B0)
S[2]=S[2].subs(C(0)==C0)
S
```

```python
# Quin valor tindrà D(t) en l'infinit quan A(t) sigui 0 en l'infinit?
sa=solve(S[0]==0,D)
sa
```

```python
# A quin valor de r correspon aquest valor de D(t) trobat anteriorment si volem que B(t) també valgui 0 en 
# l'infinit?
solve(S[1].subs(D(t)==sa[0].rhs())==0,r)
```

```python
# Per tant:
# Si r<10 ==> B0<40+10=50 el reactiu limitant seguirà sent B.
# Si r=10 ==> B0=40+10=50 s'esgotaran (en un temps infinit) els dos reactius A i B.
# Si r>10 ==> B0>40+10=50 el nou reactiu limitant serà A i per tant sobrarà compost de B per reaccionar.

# En els tres casos sobrarà sempre compost de C per reaccionar.
```
