```python
# Funció per a calcular i graficar les separatrius de punts de sella.

def separatrius(eq1,eq2,veps,px,py,k1,k2,h,stp): 
    # eq1 i eq2 són les parts de la dreta de les equacions que defineixen el sistema d'equacions diferencials
    # "veps" és una de les components de la sortida que et dona el sagemath quan calculem els veps d'alguna mariu.
    # (px,py) és el punt de sella
    # k1, k2 , h i stp són paràmetres per ajustar millor les separatrius en el gràfic.
    separ_plot=[]
    if veps[0]>0:
        sol1=desolve_system_rk4([eq1,eq2],[x1,y1],ics=[0,px+h*veps[1][0][0],py+h*veps[1][0][1]],ivar=t,step=stp,end_points=[px+k1,py+k1])
        sol11=[[j,k] for i,j,k in sol1]
        sol2=desolve_system_rk4([eq1,eq2],[x1,y1],ics=[0,px-h*veps[1][0][0],py-h*veps[1][0][1]],ivar=t,step=stp,end_points=[px+k2,py+k2])
        sol22=[[j,k] for i,j,k in sol2]
        separ_plot.append(list_plot(sol11,color='red'))
        separ_plot.append(list_plot(sol22,color='red'))
    elif veps[0]<=0:
        sol1=desolve_system_rk4([eq1,eq2],[x1,y1],ics=[0,px+h*veps[1][0][0],py+h*veps[1][0][1]],ivar=t,step=stp,end_points=[px-k1,py-k1])
        sol11=[[j,k] for i,j,k in sol1]
        sol2=desolve_system_rk4([eq1,eq2],[x1,y1],ics=[0,px-h*veps[1][0][0],py-h*veps[1][0][1]],ivar=t,step=stp,end_points=[px-k2,py-k2])
        sol22=[[j,k] for i,j,k in sol2]
        separ_plot.append(list_plot(sol11,color='blue'))
        separ_plot.append(list_plot(sol22,color='blue'))
    return sum(separ_plot)
```

```python
# Funció per a calcular òrbites i graficar-les.

def orbites(eq1,eq2,x0,y0,k1,k2,stp):
    # eq1 i eq2 són les parts de la dreta de les equacions que defineixen el sistema d'equacions diferencial.
    # (x0,y0) són les condicions inicials del sistema
    # k1 < k2 són paràmetres per ajustar millor el gràfic.
    sol1=desolve_system_rk4([eq1,eq2],[x1,y1],ics=[0,x0,y0],ivar=t,step=stp,end_points=[k1,k2])
    sol11=[[j,k] for i,j,k in sol1]
    p=list_plot(sol11,color="green")
    return p
```
